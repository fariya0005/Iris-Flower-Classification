```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# Load the Iris dataset
iris_data = load_iris()
X = iris_data.data
y = iris_data.target

# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Visualize the data with K-means clustering
colors = ["g.", "r.", "c."]
for i in range(len(X)):
    plt.plot(X[i][0], X[i][1], colors[labels[i]], markersize=5)

plt.scatter(centroids[:, 0], centroids[:, 1], marker="x", s=50, zorder=10)
plt.title("Prediction of Classification with K MEANS")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()

# Visualize the actual data
df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
df["species"] = iris_data.target_names[y]
colors2 = ["r.", "g.", "c."]
for i in range(len(df)):
    plt.plot(df.iloc[i]["sepal length (cm)"], df.iloc[i]["sepal width (cm)"], colors2[y[i]], markersize=5)

plt.title("Classification according to actual value")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
# Visualize the clusters
plt.figure(figsize=(10, 6))
print(iris_df.columns)
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='Cluster', data=iris_df, palette='viridis', legend='full')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='*', s=300, c='red', label='Centroids')
plt.title('K-means Clustering of Iris Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.show()
```
